# C. Transform the COVID Data

1. Parse the `date` column in `covid_df` to datetime objects.
2. Add a column for the year and month.
3. Group by the new year-month column and `fips` (or a combination of `state` and `county`), and sum the cases and deaths over each month. Since the data seems to be cumulative, you'll want the last day of each month for each county.
4. Create the `county_id` column which matches the `fips`.
5. Finally, construct the table with the columns ["County", "Month", "#cases", "#deaths"] for the specific counties and months you're interested in.

```python
# Convert the 'date' column to datetime
covid_df['date'] = pd.to_datetime(covid_df['date'])

# Create year-month column for grouping by month
covid_df['year_month'] = covid_df['date'].dt.to_period('M')

# Get last day entries for each month for each county which represent the cumulative number for the month
covid_monthly = covid_df.sort_values(['fips', 'date']).groupby(['year_month', 'fips']).last().reset_index()

covid_monthly
```

```python
# Drop unnecessary columns
covid_monthly = covid_monthly[['fips', 'year_month', 'cases', 'deaths']]

# Rename the year_month back to the 'date' and 'fips' to 'county_id'
covid_monthly = covid_monthly.rename(columns={'year_month': 'Month', 'fips': 'county_id'})

# Add a county name column from covid_df to covid_monthly
covid_monthly = covid_monthly.merge(covid_df[['fips', 'county', 'state']].drop_duplicates(), left_on='county_id', right_on='fips', how='left')
covid_monthly['County'] = covid_monthly['county'] + ', ' + covid_monthly['state']

# Drop the additional 'fips', 'county', and 'state' columns
covid_monthly = covid_monthly.drop(['fips', 'county', 'state'], axis=1)

# Reorder columns
covid_monthly = covid_monthly[['County', 'county_id', 'Month', 'cases', 'deaths']]

# Convert the Month period to string format
covid_monthly['Month'] = covid_monthly['Month'].dt.strftime('%B %Y').astype(str)

covid_monthly.head()
```

```python
counties = ["Malheur, Oregon"]
months = ["August 2020", "January 2021", "February 2021"]

# Now find the required data for the specified counties and months
specified_data = covid_monthly[
    (covid_monthly['County'].isin(counties)) & (covid_monthly['Month'].isin(months))
][['County', 'Month', 'cases', 'deaths']]

specified_data
```

# D. Integrate COVID Data with ACS Data

To do this:

1. Transform the COVID Data
2. Fill the Required Table for Specific Counties

```python
# Group by 'county' and 'state' and sum up the 'cases' and 'deaths'
covid_summary = covid_df.groupby(['county', 'state'], as_index=False).agg({
    'cases': 'sum',
    'deaths': 'sum'
})

# Cleanup: renaming for consistency with request
covid_summary.rename(columns={'county': 'County', 'state': 'State'}, inplace=True)

# For COVID data, concatenate County and State to form a unique identifier
covid_summary['ID'] = covid_summary['County'] + ', ' + covid_summary['State']
covid_summary['ID'] = covid_summary['ID'].str.replace(' County', '').str.strip()

# Do the same for the census DataFrame
census_df['ID'] = census_df['County'] + ', ' + census_df['State']

# Group by 'State' and 'County' and calculate the necessary statistics
census_summary = census_df.groupby(['State', 'County']).agg({
    'TotalPop': 'sum',
    'Poverty': 'mean',
    'Income': 'mean',
    'IncomePerCap': 'mean',
    'Employed': 'sum',
    'Unemployment': 'mean',
    'White': 'mean',
    'Black': 'mean',
    'Hispanic': 'mean',
    'Asian': 'mean',
    'Drive': 'mean',
    'Carpool': 'mean',
    'Transit': 'mean',
    'Walk': 'mean',
    'OtherTransp': 'mean',
    'WorkAtHome': 'mean',
    'MeanCommute': 'mean',
}).reset_index()

census_summary.columns = [
    'State', 'County', 'TotalPopulation', 'AveragePoverty', 'AverageIncome',
    'AverageIncomePerCap', 'TotalEmployed', 'AverageUnemployment', 'AverageWhite',
    'AverageBlack', 'AverageHispanic', 'AverageAsian', 'AverageDrive',
    'AverageCarpool', 'AverageTransit', 'AverageWalk', 'AverageOtherTransp',
    'AverageWorkAtHome', 'AverageMeanCommute'
]

# Cleanup: renaming for consistency with request
census_summary.rename(columns={'TotalPopulation': 'Population', 'AveragePoverty': 'Poverty%', 'AverageIncomePerCap': 'PerCapitaIncome'}, inplace=True)

# Add the unique identifier
census_summary['ID'] = census_summary['County'] + ', ' + census_summary['State']
census_summary['ID'] = census_summary['ID'].str.replace(' County', '').str.strip()

census_summary = census_summary.drop(['County', 'State'], axis=1)

# Merge covid_summary with census_summary on 'ID'
# COVID_summary = pd.merge(covid_summary, census_summary[['ID', 'Population', 'Poverty%', 'PerCapitaIncome']], on='ID', how='inner')
COVID_summary = pd.merge(covid_summary, census_summary, on='ID', how='inner')

# Calculate per 100K metrics
COVID_summary['TotalCasesPer100K'] = COVID_summary['cases'] * 100000 / COVID_summary['Population']
COVID_summary['TotalDeathsPer100K'] = COVID_summary['deaths'] * 100000 / COVID_summary['Population']

COVID_summary.head()
```
